```python
import pandas as pd

# Load the correct dataset as df
df = pd.read_csv('/content/Bengaluru_House_Data (2).csv')

print("Missing values in 'society' column:")
print(df['society'].isnull().sum())

print("\nData type of 'society' column:")
print(df['society'].dtype)

print("\nUnique values and their frequencies in 'society' column:")
print(df['society'].value_counts(dropna=False))
```

```python
df['society'] = df['society'].fillna('No Society')

print("Missing values in 'society' column after imputation:")
print(df['society'].isnull().sum())

print("\nUnique values and their frequencies in 'society' column after imputation:")
print(df['society'].value_counts(dropna=False))
```

```python
print("Descriptive statistics for 'price' column:")
print(df['price'].describe())

print("\nData type of 'price' column:")
print(df['price'].dtype)
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]

print(f"\nNumber of identified outliers: {len(outliers)}")
print("\nIdentified outlier rows based on 'price' column:")
print(outliers[['price']].head()) # Displaying first few outlier rows for brevity, focusing on price
```

```python
cleaned_df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

print("Shape of the original DataFrame:", df.shape)
print("Shape of the cleaned DataFrame after removing price outliers:", cleaned_df.shape)
print(f"Number of rows removed: {df.shape[0] - cleaned_df.shape[0]}")
```
